```python
def compute_expected_value(series):
    prob = series.sum() / len(series)
    
    return prob * series.sum() + (1 - prob) * (len(series) - series.sum())
```

```python
def compute_entropy(series):
    prob = series.sum() / len(series)
    
    return  - (prob * np.log2(prob) + (1 - prob) * np.log2(1 - prob))
```

```python
item = df.sample(1, random_state=41).squeeze()
item
```

```python
df.apply(compute_entropy, axis=0).sort_values()
```

```python
def greedy_search(item, candidates, metric_computer):
    
    print("Using {} as metric to minimize!".format(metric_computer.__name__))
    
    df = candidates.copy()
    
    # number of questions asked
    count = 0
    
    while len(df) > 1:
        # search for feature with lowest expected value
        expected_values = df.apply(metric_computer, axis=0)
        pivot = expected_values.argmin()

        # check feature against item
        pivot_value = item[pivot]

        print("pivot: {}".format(pivot))

        df = df[df[pivot] == pivot_value]
        
        count += 1
        
        print("Remaining candidates: {}".format(len(df)))
        
        # safety valve
        if count > 10:
            break
    
    return df.squeeze(), count
```

```python
solution, count = greedy_search(item, df, compute_expected_value)
assert solution.name == item.name
print('Finding solution took {} questions.'.format(count))
```

```python
solution, count = greedy_search(item, df, compute_entropy)
assert solution.name == item.name
print('Finding solution took {} questions.'.format(count))
```

```python
results = {}
for character in characters:
    solution, count = greedy_search(df.loc[character, :], df, compute_expected_value)
    results[character] = count
```

```python
values = list(results.values())
sum(values) / len(values) 
```
