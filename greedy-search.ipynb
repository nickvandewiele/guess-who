# Import Data

```python
df = pd.DataFrame(data=features_data, columns=features, index=characters)
df
```

# Test greedy search using expected value

```python
all_results = {}
```

```python
results = {}
for character in characters:
    solution, count = greedy_search(df.loc[character, :], df, expected_value)
    results[character] = count
```

```python
values = list(results.values())
sum(values) / len(values) 

all_results['greedy, expected value'] = sum(values) / len(values)
```
